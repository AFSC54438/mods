```python
from googleapiclient.discovery import build
# from json import dumps

def get_playlist_video_links(api_key, playlist_id):
    # Initialize the YouTube API client
    youtube = build('youtube', 'v3', developerKey=api_key)
    
    video_links = {}
    next_page_token = None
    
    while True:
        # Request a page of playlist items
        request = youtube.playlistItems().list(
            part='snippet',
            playlistId=playlist_id,
            maxResults=50,  # Max allowable per request
            pageToken=next_page_token
        )
        response = request.execute()
        
        # Extract video IDs and format URLs
        for item in response.get('items', []):
            video_id = item['snippet']['resourceId']['videoId']
            video_url = f"https://www.youtube.com/watch?v={video_id}"
            video_links[video_url] = item['snippet']['title']
            
        # Check if there is another page of videos
        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break  # Exit loop if no more pages exist
            
    return video_links
```

```python
def get_playlist_video_ids(api_key, playlist_id):
    # Initialize the YouTube API client
    youtube = build('youtube', 'v3', developerKey=api_key)
    
    video_links = []
    next_page_token = None
    
    while True:
        # Request a page of playlist items
        request = youtube.playlistItems().list(
            part='snippet',
            playlistId=playlist_id,
            maxResults=50,  # Max allowable per request
            pageToken=next_page_token
        )
        response = request.execute()
        
        # Extract video IDs and format URLs
        for item in response.get('items', []):
            vid = {}
            vid["id"] = item['snippet']['resourceId']['videoId']
            vid["title"] = item['snippet']['title']
            video_links.append(vid)
            
        # Check if there is another page of videos
        next_page_token = response.get('nextPageToken')
        if not next_page_token:
            break  # Exit loop if no more pages exist
            
    return video_links
    
```

```python
# Example Configuration
PLAYLIST_ID = "PLdx_WUzWEp4uRRo34Nco-rCX9f8nIScpp"

# Run the function
PLAYLIST_VIDEOS = get_playlist_video_ids(api_key, PLAYLIST_ID)
PLAYLIST_VIDEOS.reverse()

playlist_len = len(PLAYLIST_VIDEOS)


print(f"Total videos found: {playlist_len}")
for count, i in enumerate(PLAYLIST_VIDEOS):
    print(f"{count}. {i["id"]} | {i["title"]}")
```
